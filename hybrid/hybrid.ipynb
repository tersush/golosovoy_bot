```python
open_prompt = open("prompt_golosovoy_bot.txt")
prompt = open_prompt.read()
```

```python
import requests
from dotenv import load_dotenv
import os

load_dotenv("key_server.env")
API_TOKEN = os.getenv("VLLM_API_TOKEN")
if not API_TOKEN:
    raise ValueError("‚ùå –¢–æ–∫–µ–Ω –Ω–µ –Ω–∞–π–¥–µ–Ω!")

url = "http://gojdaforever.ru:9696/v1/chat/completions"
headers = {"Authorization": f"Bearer {API_TOKEN}"}

# –ò—Å—Ç–æ—Ä–∏—è –¥–∏–∞–ª–æ–≥–∞ ‚Äî –æ—Ç–¥–µ–ª—å–Ω–æ –æ—Ç `data`
messages = []

messages=[{
    "role": "system", 
    "content": prompt}]

data = {
        "model": "qwen3-14b-awq",
        "messages": messages,  # ‚Üê –ø–µ—Ä–µ–¥–∞—ë–º –≤—Å—é –∏—Å—Ç–æ—Ä–∏—é
        "max_tokens": 1000,
        "temperature": 0.7
    }

    # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º –∑–∞–ø—Ä–æ—Å
response = requests.post(url, headers=headers, json=data)

result = response.json()
assistant_reply = result["choices"][0]["message"]["content"]

    # –í—ã–≤–æ–¥–∏–º –æ—Ç–≤–µ—Ç
print(f"\nü§ñ –ê—Å—Å–∏—Å—Ç–µ–Ω—Ç: {assistant_reply.strip()}")

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ—Ç–≤–µ—Ç –∞—Å—Å–∏—Å—Ç–µ–Ω—Ç–∞ –≤ –∏—Å—Ç–æ—Ä–∏—é
messages.append({"role": "assistant", "content": assistant_reply})

while True:
    user_input = input().strip()
    if user_input.lower() in ("–∑–∞–≤–µ—Ä—à–∏—Ç—å —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫—É", "–∑–∞–∫–æ–Ω—á–∏—Ç—å", "–≤—ã–π—Ç–∏"):
        break

    # –î–æ–±–∞–≤–ª—è–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    messages.append({"role": "user", "content": user_input})

    # –§–æ—Ä–º–∏—Ä—É–µ–º —Ç–µ–ª–æ –∑–∞–ø—Ä–æ—Å–∞
    data = {
        "model": "qwen3-14b-awq",
        "messages": messages,  # ‚Üê –ø–µ—Ä–µ–¥–∞—ë–º –≤—Å—é –∏—Å—Ç–æ—Ä–∏—é
        "max_tokens": 1000,
        "temperature": 0.7
    }

    # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º –∑–∞–ø—Ä–æ—Å
    response = requests.post(url, headers=headers, json=data)
    if response.status_code != 200:
        print(f"‚ùå –û—à–∏–±–∫–∞: {response.status_code} ‚Äî {response.text}")
        break

    result = response.json()
    assistant_reply = result["choices"][0]["message"]["content"]

    # –í—ã–≤–æ–¥–∏–º –æ—Ç–≤–µ—Ç
    print(f"\nü§ñ –ê—Å—Å–∏—Å—Ç–µ–Ω—Ç: {assistant_reply.strip()}")

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ—Ç–≤–µ—Ç –∞—Å—Å–∏—Å—Ç–µ–Ω—Ç–∞ –≤ –∏—Å—Ç–æ—Ä–∏—é
    messages.append({"role": "assistant", "content": assistant_reply})
```
